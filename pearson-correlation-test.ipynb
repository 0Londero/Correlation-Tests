```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import kagglehub
```

```python
path = kagglehub.dataset_download("tawfikelmetwally/advertising-dataset")
```

```python
df = pd.read_csv(path + r'\Advertising.csv')
```

### RESEARCH QUESTIONS:
1- Is there a significant relationship between TV advertising and sales?

![Rule of Thumb](images/1.png)

```python
def evaluating_strength(correlation_coefficient):
    abs_r = abs(correlation_coefficient)
    if correlation_coefficient == 0:
        return "No correlation"
    elif 0 < abs_r < 0.3:
        return "Weak correlation (linear relationship is very weak or negligible)"
    elif 0.3 <= abs_r < 0.5:
        return "Moderate correlation"
    elif 0.5 <= abs_r < 0.7:
        return "Strong correlation"
    elif 0.7 <= abs_r < 1:
        return "Very strong correlation"
    elif abs_r == 1:
        return "Perfect correlation"
    else:
        return "Invalid value for r"
```

```python
stats.pearsonr(df['TV'], df['Sales'])
```

```python
# statistic = Correlation Coefficient (r)
# pvalue = probability value
r1, p1 = stats.pearsonr(df['TV'], df['Sales'])

print (f"\nPearson correlation coefficient(r)    -> {r1}\n{evaluating_strength(r1)}\nPvalue     ->{p1:.4f}")

print (f"\n-- This is the one that has the strongest correlation with sales")
```

```python
# kws = keywords arguments
sns.regplot(x='TV', y='Sales', data=df,
            line_kws={'color': '#6cd4c5'},
            scatter_kws={'alpha': 0.6, 'color': '#e27c7c'})

plt.title(f'Pearson R = {r1:.2f} | p-value = {p1:.2f}')
plt.xlabel('TV Advertising Budget $USD')
plt.ylabel('Sales $USD')
plt.grid(True)
plt.show()
```
